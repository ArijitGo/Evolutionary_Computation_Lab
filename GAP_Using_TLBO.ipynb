```python
import numpy as np

# ------------------------------------------------------------
# Helper functions (evaluation, fitness, repair, greedy)
# ------------------------------------------------------------
def evaluate_assignment(assign, cost, resource, capacity):
    """Return total cost and total resource violation for an integer assignment."""
    total_cost = np.sum(cost[assign, np.arange(len(assign))])
    used = np.zeros(len(capacity), dtype=int)
    np.add.at(used, assign, resource[assign, np.arange(len(assign))])
    violation = np.sum(np.maximum(0, used - capacity))
    return total_cost, violation

def compute_fitness(x, cost, resource, capacity, penalty):
    """Fitness of a real‑coded solution x (maximisation: fitness = total_cost - penalty * violation)."""
    assign = np.floor(x).astype(int)
    total_cost = np.sum(cost[assign, np.arange(len(assign))])
    used = np.zeros(len(capacity), dtype=int)
    np.add.at(used, assign, resource[assign, np.arange(len(assign))])
    violation = np.sum(np.maximum(0, used - capacity))
    fitness = total_cost - penalty * violation
    return fitness, total_cost, violation

def greedy_feasible_assign_multiple(cost, resource, capacity, n_trials=20):
    """
    Try to construct a feasible assignment using a greedy heuristic with random task orders.
    Returns a feasible assignment (numpy array) if found, otherwise None.
    """
    n_agents, n_tasks = cost.shape
    for _ in range(n_trials):
        assign = np.zeros(n_tasks, dtype=int) - 1
        remaining = capacity.copy()
        tasks = list(range(n_tasks))
        np.random.shuffle(tasks)
        feasible = True
        for t in tasks:
            feasible_agents = [a for a in range(n_agents) if resource[a, t] <= remaining[a]]
            if not feasible_agents:
                feasible = False
                break
            best = max(feasible_agents, key=lambda a: cost[a, t])
            assign[t] = best
            remaining[best] -= resource[best, t]
        if feasible:
            return assign
    return None

def repair_assignment(assign, cost, resource, capacity, max_iter=5000):
    """
    Advanced repair of an infeasible assignment.
    Tries single moves, then swaps, and falls back to multi‑trial greedy.
    Returns a feasible assignment and a success flag.
    """
    n_agents, n_tasks = cost.shape
    # Compute current used resources
    used = np.zeros(n_agents, dtype=int)
    for a in range(n_agents):
        used[a] = np.sum(resource[a, assign == a])

    # Early exit if already feasible
    if np.all(used <= capacity):
        return assign, True

    new_assign = assign.copy()
    used = np.zeros(n_agents, dtype=int)
    for a in range(n_agents):
        used[a] = np.sum(resource[a, new_assign == a])

    for _ in range(max_iter):
        overloaded = [a for a in range(n_agents) if used[a] > capacity[a]]
        if not overloaded:
            break   # feasible now

        # Work on the most overloaded agent
        a = max(overloaded, key=lambda a: used[a] - capacity[a])
        tasks_on_a = np.where(new_assign == a)[0]
        np.random.shuffle(tasks_on_a)

        # ---- Attempt single moves ----
        best_gain = -np.inf
        best_task = None
        best_target = None
        for i in tasks_on_a:
            for b in range(n_agents):
                if b == a:
                    continue
                if used[b] + resource[b, i] <= capacity[b]:
                    gain = cost[b, i] - cost[a, i]
                    if gain > best_gain:
                        best_gain = gain
                        best_task = i
                        best_target = b

        if best_task is not None:
            # Perform the best move
            new_assign[best_task] = best_target
            used[a] -= resource[a, best_task]
            used[best_target] += resource[best_target, best_task]
            continue

        # ---- If no move works, try swaps ----
        best_swap_gain = -np.inf
        best_swap = None   # (i_on_a, j_on_b, b)
        for i in tasks_on_a:
            for b in range(n_agents):
                if b == a:
                    continue
                tasks_on_b = np.where(new_assign == b)[0]
                for j in tasks_on_b:
                    # Check if swapping i and j keeps both agents within capacity
                    new_used_a = used[a] - resource[a, i] + resource[a, j]
                    new_used_b = used[b] - resource[b, j] + resource[b, i]
                    if new_used_a <= capacity[a] and new_used_b <= capacity[b]:
                        gain = (cost[b, i] + cost[a, j]) - (cost[a, i] + cost[b, j])
                        if gain > best_swap_gain:
                            best_swap_gain = gain
                            best_swap = (i, j, b)

        if best_swap is not None:
            i, j, b = best_swap
            # Perform swap
            new_assign[i] = b
            new_assign[j] = a
            used[a] = used[a] - resource[a, i] + resource[a, j]
            used[b] = used[b] - resource[b, j] + resource[b, i]
            continue

        # If we reach here, neither moves nor swaps could fix this overloaded agent.
        # Fall back to multi‑trial greedy.
        greedy_assign = greedy_feasible_assign_multiple(cost, resource, capacity, n_trials=20)
        if greedy_assign is not None:
            return greedy_assign, True
        else:
            return assign, False

    feasible = np.all(used <= capacity)
    return new_assign, feasible

# ------------------------------------------------------------
# TLBO algorithm exactly following the pseudocode (maximisation)
# ------------------------------------------------------------
def tlbo_gap_pseudocode(cost, resource, capacity, pop_size=50, max_iter=500,
                        penalty=10000, minx=0, maxx=None, seed=None):
    """
    Implements the TLBO pseudocode exactly (maximisation version).
    Returns a feasible assignment.
    """
    if seed is not None:
        np.random.seed(seed)

    n_agents, n_tasks = np.array(cost).shape
    cost_np = np.array(cost)
    resource_np = np.array(resource)
    capacity_np = np.array(capacity)

    # Ensure maxx is less than n_agents to avoid index n_agents after floor
    if maxx is None or maxx >= n_agents:
        maxx = n_agents - 1e-10   # safe upper bound

    # Step 1: Randomly initialize class of N students
    class_pos = [np.random.uniform(minx, maxx, n_tasks) for _ in range(pop_size)]
    class_fit = []
    for x in class_pos:
        f, _, _ = compute_fitness(x, cost_np, resource_np, capacity_np, penalty)
        class_fit.append(f)

    best_idx = np.argmax(class_fit)
    best_sol = class_pos[best_idx].copy()
    best_fit = class_fit[best_idx]

    # Step 2: Iterate
    for it in range(max_iter):
        # For each student (sequential update)
        for i in range(pop_size):
            # ---------- Teaching phase ----------
            teacher = class_pos[np.argmax(class_fit)]   # teacher = best student
            mean = np.mean(class_pos, axis=0)
            TF = np.random.choice([1, 2])
            r = np.random.rand()
            new_x = class_pos[i] + r * (teacher - TF * mean)

            # Clip to [minx, maxx]
            new_x = np.clip(new_x, minx, maxx)

            # Compute fitness
            new_fit, _, _ = compute_fitness(new_x, cost_np, resource_np, capacity_np, penalty)

            # Greedy selection (maximisation)
            if new_fit > class_fit[i]:
                class_pos[i] = new_x
                class_fit[i] = new_fit
                if new_fit > best_fit:
                    best_fit = new_fit
                    best_sol = new_x.copy()

            # ---------- Learning phase ----------
            # Choose random partner j ≠ i
            j = i
            while j == i:
                j = np.random.randint(pop_size)

            r = np.random.rand()
            if class_fit[i] > class_fit[j]:
                new_x = class_pos[i] + r * (class_pos[i] - class_pos[j])
            else:
                new_x = class_pos[i] + r * (class_pos[j] - class_pos[i])

            new_x = np.clip(new_x, minx, maxx)
            new_fit, _, _ = compute_fitness(new_x, cost_np, resource_np, capacity_np, penalty)

            if new_fit > class_fit[i]:
                class_pos[i] = new_x
                class_fit[i] = new_fit
                if new_fit > best_fit:
                    best_fit = new_fit
                    best_sol = new_x.copy()

        if it % 50 == 0:
            print(f"Iteration {it:4d}, best fitness = {best_fit:.2f}")

    # Decode best solution and repair to guarantee feasibility
    best_assign = np.floor(best_sol).astype(int)
    # Extra safety: clip assignments to valid range
    best_assign = np.clip(best_assign, 0, n_agents - 1)

    best_cost, violation = evaluate_assignment(best_assign, cost_np, resource_np, capacity_np)

    if violation > 0:
        print("\nBest solution infeasible. Repairing...")
        repaired_assign, feasible = repair_assignment(best_assign, cost_np, resource_np, capacity_np)
        if feasible:
            best_assign = repaired_assign
            best_cost, violation = evaluate_assignment(best_assign, cost_np, resource_np, capacity_np)
            print("Repair successful.")
        else:
            print("Repair failed. Falling back to greedy construction.")
            greedy_assign = greedy_feasible_assign_multiple(cost_np, resource_np, capacity_np, n_trials=50)
            if greedy_assign is not None:
                best_assign = greedy_assign
                best_cost, violation = evaluate_assignment(best_assign, cost_np, resource_np, capacity_np)
                print("Greedy construction succeeded.")
            else:
                print("ERROR: No feasible assignment exists.")

    # Final check
    final_cost, final_viol = evaluate_assignment(best_assign, cost_np, resource_np, capacity_np)
    if final_viol > 0:
        print("WARNING: Final solution is infeasible!")
    return best_assign, final_cost

if __name__ == "__main__":
    assign, total_cost = tlbo_gap_pseudocode(
        cost, resource, capacity,
        pop_size=100, max_iter=100, penalty=10000, seed=42
    )

    print("\n" + "="*50)
    print("Final feasible solution:")
    print("Assignment (agent per task):", assign)
    print(f"Total cost: {total_cost}")

    # Double-check feasibility
    _, v = evaluate_assignment(assign, np.array(cost), np.array(resource), np.array(capacity))
    if v == 0:
        print("The solution is feasible (zero violation).")
    else:
        print(f"WARNING: solution has violation {v} – this should not happen.")
```
