```python
import numpy as np
import random
from typing import List, Tuple

class AssignmentProblemGA:
    def __init__(self, cost: List[List[int]], resource: List[List[int]], capacity: List[int]):
        self.cost = np.array(cost)
        self.resource = np.array(resource)
        self.capacity = np.array(capacity)
        self.n_agents = len(cost)
        self.n_tasks = len(cost[0])
        
        # Each task needs ceil(log2(n_agents)) bits
        self.bits_per_task = (self.n_agents - 1).bit_length()
        self.chromosome_length = self.n_tasks * self.bits_per_task
        
        # GA parameters
        self.pop_size = 100
        self.max_generations = 500
        self.crossover_rate = 0.8
        self.mutation_rate = 0.004
        self.elitism_count = 2
        
    def _binary_to_agent(self, binary_str: str) -> int:
        """Convert binary string to agent index (0 to n_agents-1)"""
        agent_idx = int(binary_str, 2)
        return agent_idx % self.n_agents
    
    def _decode_chromosome(self, chromosome: str) -> List[int]:
        """Decode binary chromosome to assignment list"""
        assignments = []
        for i in range(self.n_tasks):
            start = i * self.bits_per_task
            end = start + self.bits_per_task
            binary_str = chromosome[start:end]
            assignments.append(self._binary_to_agent(binary_str))
        return assignments
    
    def _evaluate_fitness(self, chromosome: str) -> float:
        """Evaluate fitness of a chromosome"""
        assignments = self._decode_chromosome(chromosome)
        total_cost = 0
        agent_resources = np.zeros(self.n_agents)
        
        # Calculate total cost and resource usage
        for task, agent in enumerate(assignments):
            total_cost += self.cost[agent][task]
            agent_resources[agent] += self.resource[agent][task]
        
        # Check capacity constraints and apply penalty
        penalty = 0
        for agent in range(self.n_agents):
            if agent_resources[agent] > self.capacity[agent]:
                # Quadratic penalty for constraint violation
                penalty += 1000 * (agent_resources[agent] - self.capacity[agent]) ** 2
        
        return total_cost - penalty
    
    def _create_individual(self) -> str:
        """Create a random individual"""
        return ''.join(str(random.randint(0, 1)) for _ in range(self.chromosome_length))
    
    def _select_parents(self, population: List[str], fitnesses: List[float]) -> Tuple[str, str]:
        """Tournament selection"""
        tournament_size = 3
        selected = []
        
        for _ in range(2):
            tournament_indices = random.sample(range(len(population)), tournament_size)
            tournament_fitness = [fitnesses[i] for i in tournament_indices]
            winner_idx = tournament_indices[np.argmax(tournament_fitness)]
            selected.append(population[winner_idx])
        
        return selected[0], selected[1]
    
    def _crossover(self, parent1: str, parent2: str) -> Tuple[str, str]:
        """Single-point crossover"""
        if random.random() > self.crossover_rate:
            return parent1, parent2
        
        point = random.randint(1, self.chromosome_length - 1)
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
        
        return child1, child2
    
    def _mutate(self, chromosome: str) -> str:
        """Bit-flip mutation"""
        mutated = list(chromosome)
        for i in range(len(mutated)):
            if random.random() < self.mutation_rate:
                mutated[i] = '0' if mutated[i] == '1' else '1'
        return ''.join(mutated)
    
    def solve(self) -> Tuple[List[int], float, List[int]]:
        """Main GA algorithm"""
        # Initialize population
        population = [self._create_individual() for _ in range(self.pop_size)]
        best_fitness_history = []
        
        for generation in range(self.max_generations):
            # Evaluate fitness
            fitnesses = [self._evaluate_fitness(ind) for ind in population]
            
            # Track best solution
            best_idx = np.argmax(fitnesses)
            best_fitness = fitnesses[best_idx]
            best_fitness_history.append(best_fitness)
            
            # Create new population with elitism
            sorted_indices = np.argsort(fitnesses)[::-1]
            new_population = [population[i] for i in sorted_indices[:self.elitism_count]]
            
            # Generate offspring
            while len(new_population) < self.pop_size:
                parent1, parent2 = self._select_parents(population, fitnesses)
                child1, child2 = self._crossover(parent1, parent2)
                child1 = self._mutate(child1)
                child2 = self._mutate(child2)
                new_population.extend([child1, child2])
            
            population = new_population[:self.pop_size]
            
            # Print progress
            if generation % 50 == 0:
                print(f"Generation {generation}: Best fitness = {best_fitness}")
        
        # Get best solution
        final_fitnesses = [self._evaluate_fitness(ind) for ind in population]
        best_idx = np.argmax(final_fitnesses)
        best_chromosome = population[best_idx]
        best_assignments = self._decode_chromosome(best_chromosome)
        best_cost = self._evaluate_fitness(best_chromosome)
        
        # Calculate resource usage per agent
        agent_resources = np.zeros(self.n_agents)
        for task, agent in enumerate(best_assignments):
            agent_resources[agent] += self.resource[agent][task]
        
        return best_assignments, best_cost, agent_resources.tolist()

# Run the algorithm
if __name__ == "__main__":
    # Initialize with provided data
    ga_solver = AssignmentProblemGA(cost, resource, capacity)
    
    # Solve the problem
    assignments, total_cost, resource_usage = ga_solver.solve()
    
    print(f"\nOptimal Assignment:")
    for task, agent in enumerate(assignments):
        print(f"  Task {task+1:2d} -> Agent {agent+1}")
    
    print(f"\nTotal Cost: {total_cost}")
    print("\nResource Usage per Agent:")
    for i, usage in enumerate(resource_usage):
        print(f"  Agent {i+1}: {usage:.2f}/{capacity[i]} ({usage/capacity[i]*100:.1f}%)")
    
    # Check for constraint violations
    feasible = all(usage <= cap for usage, cap in zip(resource_usage, capacity))
    print(f"\nSolution Feasible: {'Yes' if feasible else 'No'}")
```
