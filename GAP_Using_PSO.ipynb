```python
import numpy as np
import time

# =====================================================
# SET RANDOM SEED
# =====================================================
np.random.seed(42)

# =====================================================
# DATA – REPLACE WITH YOUR ACTUAL ARRAYS
# =====================================================
cost = np.array(cost)          # shape: (10, 60)
resource = np.array(resource)  # shape: (10, 60)
capacity = np.array(capacity)  # shape: (10,)

n_machines = cost.shape[0]
n_jobs = cost.shape[1]

# =====================================================
# PSO PARAMETERS
# =====================================================
pop_size = 100
generations = 300
w_max = 0.9
w_min = 0.4
c1 = 2.0
c2 = 2.0
Xl, Xu = 0.0, 1.0
Vmax = 0.2 * (Xu - Xl)
penalty_factor = 100000

# =====================================================
# DECODE (REAL → ASSIGNMENT)
# =====================================================
def decode(positions):
    if positions.ndim == 1:
        positions = positions.reshape(1, -1)
    assignment = (positions * n_machines).astype(int)
    return np.clip(assignment, 0, n_machines - 1)

# =====================================================
# FITNESS FUNCTION (MINIMIZATION)
# =====================================================
def fitness(positions):
    if positions.ndim == 1:
        positions = positions.reshape(1, -1)
    assignment = decode(positions)
    rows = np.arange(n_jobs)
    total_cost = np.sum(cost[assignment, rows], axis=1)
    total_resource = np.zeros((positions.shape[0], n_machines))
    np.add.at(total_resource,
              (np.arange(positions.shape[0])[:, None], assignment),
              resource[assignment, rows])
    violation = np.maximum(total_resource - capacity, 0)
    penalty = violation.sum(axis=1) * penalty_factor
    return total_cost + penalty

# =====================================================
# PHASE 1: ITERATIVE MOVE REPAIR (fixes a single assignment)
# =====================================================
def iterative_move_repair(assignment):
    """
    Takes an integer assignment (shape (n_jobs,)) and attempts to make it feasible
    by moving jobs from overloaded machines to others with spare capacity.
    Returns a feasible assignment if successful, otherwise None.
    """
    rows = np.arange(n_jobs)
    used = np.zeros(n_machines)
    np.add.at(used, assignment, resource[assignment, rows])

    # Keep trying until feasible or no moves possible
    max_iter = 100  # prevent infinite loops
    for _ in range(max_iter):
        if np.all(used <= capacity):
            return assignment.copy()  # feasible

        # Process one overloaded machine at a time
        overloaded = np.where(used > capacity)[0]
        for m in overloaded:
            jobs_on_m = np.where(assignment == m)[0]
            # Sort jobs by resource consumption (descending) – moving larger jobs helps more
            job_resources = resource[m, jobs_on_m]
            sorted_jobs = jobs_on_m[np.argsort(job_resources)[::-1]]

            moved = False
            for job in sorted_jobs:
                res_job = resource[:, job]
                # Try to move to any other machine with enough spare capacity
                for new_m in range(n_machines):
                    if new_m == m:
                        continue
                    if used[new_m] + res_job[new_m] <= capacity[new_m]:
                        # Perform the move
                        assignment[job] = new_m
                        used[m] -= res_job[m]
                        used[new_m] += res_job[new_m]
                        moved = True
                        break
                if moved:
                    break
            # After processing this machine, break out of the overloaded loop
            # because the usage has changed; restart the outer while.
            if moved:
                break
        else:
            # No move was made on any overloaded machine – cannot improve further
            return None
    return None  # max iterations reached without becoming feasible

# =====================================================
# PHASE 2: GREEDY CONSTRUCTIVE HEURISTIC (fallback)
# =====================================================
def greedy_constructive():
    """
    Builds a feasible assignment from scratch using a greedy rule:
    sort jobs by descending resource demand on the cheapest machine,
    then assign each job to the machine with smallest cost that still has
    enough remaining capacity.
    Returns a feasible assignment (n_jobs,) or None if infeasible.
    """
    # Sort jobs by a heuristic: we use the maximum resource demand across machines
    # (descending) to place large jobs first.
    job_indices = np.arange(n_jobs)
    max_resource = np.max(resource, axis=0)  # max over machines for each job
    sorted_jobs = job_indices[np.argsort(-max_resource)]  # descending

    assignment = np.empty(n_jobs, dtype=int)
    remaining = capacity.copy()

    for job in sorted_jobs:
        # Find machines that can still accommodate this job
        feasible_machines = np.where(resource[:, job] <= remaining)[0]
        if len(feasible_machines) == 0:
            return None  # no machine can take this job → infeasible instance
        # Among feasible machines, pick the one with smallest cost
        best_machine = feasible_machines[np.argmin(cost[feasible_machines, job])]
        assignment[job] = best_machine
        remaining[best_machine] -= resource[best_machine, job]

    return assignment

# =====================================================
# INITIALIZE PARTICLES
# =====================================================
positions = np.random.rand(pop_size, n_jobs)
velocities = np.random.uniform(-Vmax, Vmax, (pop_size, n_jobs))

pbest_pos = positions.copy()
pbest_val = fitness(positions)

gbest_idx = np.argmin(pbest_val)
gbest_pos = pbest_pos[gbest_idx].copy()
gbest_val = pbest_val[gbest_idx]

# =====================================================
# PSO MAIN LOOP
# =====================================================
start_time = time.time()

for it in range(generations):
    w = w_max - (w_max - w_min) * it / (generations - 1)

    r1 = np.random.rand(pop_size, n_jobs)
    r2 = np.random.rand(pop_size, n_jobs)

    velocities = (w * velocities +
                  c1 * r1 * (pbest_pos - positions) +
                  c2 * r2 * (gbest_pos - positions))
    velocities = np.clip(velocities, -Vmax, Vmax)

    positions = positions + velocities
    positions = np.clip(positions, Xl, Xu)

    current_val = fitness(positions)

    improved = current_val < pbest_val
    pbest_pos[improved] = positions[improved]
    pbest_val[improved] = current_val[improved]

    current_best_idx = np.argmin(current_val)
    if current_val[current_best_idx] < gbest_val:
        gbest_val = current_val[current_best_idx]
        gbest_pos = positions[current_best_idx].copy()

    if it % 50 == 0:
        assignment = decode(gbest_pos)
        total_cost = cost[assignment, np.arange(n_jobs)].sum()
        print(f"Iteration {it:3d} | Actual Cost = {total_cost}")

end_time = time.time()
runtime = end_time - start_time

# =====================================================
# ENSURE FINAL SOLUTION IS FEASIBLE
# =====================================================
# Decode the global best
best_assignment = decode(gbest_pos).flatten()

# Try iterative move repair first
repaired_assignment = iterative_move_repair(best_assignment.copy())

if repaired_assignment is not None:
    final_assignment = repaired_assignment
    print("\nFeasible solution obtained via iterative move repair.")
else:
    print("\nIterative repair failed. Falling back to greedy constructive heuristic...")
    final_assignment = greedy_constructive()
    if final_assignment is None:
        raise RuntimeError("Cannot find any feasible assignment. Check total capacity vs. resource demands.")

# Compute final cost and resource usage
final_cost = cost[final_assignment, np.arange(n_jobs)].sum()
used_resource = np.zeros(n_machines)
np.add.at(used_resource, final_assignment, resource[final_assignment, np.arange(n_jobs)])

print("\n========== FINAL RESULT ==========")
print(f"Actual Total Cost: {final_cost}")
print("Capacity Used:", used_resource)
print("Capacity Limit:", capacity)
print("Capacity Violation:", np.maximum(used_resource - capacity, 0))
print("Is Feasible?:", np.all(used_resource <= capacity))
print(f"Runtime: {runtime:.2f} seconds")
print("\nFinal Assignment (job → machine):")
print(final_assignment)
```
