```python
import numpy as np

cost = np.array(cost)
resource = np.array(resource)  
capacity = np.array([79, 67, 74, 73, 67, 78, 73, 68, 73, 69])

n_machines = cost.shape[0]
n_jobs = cost.shape[1]

pop_size = 100
generations = 300
crossover_rate = 0.8
mutation_rate = 0.1

eta_c = 20
eta_m = 20

Xl = 0.0
Xu = 1.0

penalty_factor = 10000  # strong penalty

# =====================================================
# DECODE (REAL → ASSIGNMENT)
# Each job gets exactly ONE machine
# =====================================================

def decode(pop):
    # pop shape: (pop_size, n_jobs)
    assignment = (pop * n_machines).astype(int)
    return np.clip(assignment, 0, n_machines - 1)

# =====================================================
# FITNESS (MAXIMIZATION)
# =====================================================

def fitness(pop):

    assignment = decode(pop)

    total_cost = np.zeros(pop.shape[0])
    total_resource = np.zeros((pop.shape[0], n_machines))

    for i in range(pop.shape[0]):
        machines = assignment[i]
        jobs = np.arange(n_jobs)

        total_cost[i] = cost[machines, jobs].sum()

        np.add.at(
            total_resource[i],
            machines,
            resource[machines, jobs]
        )

    violation = np.maximum(total_resource - capacity, 0)
    penalty = violation.sum(axis=1) * penalty_factor

    return total_cost - penalty

# =====================================================
# TOURNAMENT SELECTION
# =====================================================

def selection(pop, fit):

    selected = []

    for _ in range(pop_size):
        i, j = np.random.randint(0, pop_size, 2)
        winner = i if fit[i] > fit[j] else j
        selected.append(pop[winner])

    return np.array(selected)

# =====================================================
# SBX CROSSOVER
# =====================================================

def sbx_crossover(pop):

    new_pop = pop.copy()

    for i in range(0, pop_size, 2):

        if np.random.rand() < crossover_rate:

            p1 = pop[i]
            p2 = pop[i+1]

            u = np.random.rand(n_jobs)

            beta = np.where(
                u <= 0.5,
                (2*u)**(1/(eta_c+1)),
                (1/(2*(1-u)))**(1/(eta_c+1))
            )

            c1 = 0.5*((1+beta)*p1 + (1-beta)*p2)
            c2 = 0.5*((1-beta)*p1 + (1+beta)*p2)

            new_pop[i] = np.clip(c1, Xl, Xu)
            new_pop[i+1] = np.clip(c2, Xl, Xu)

    return new_pop

# =====================================================
# POLYNOMIAL MUTATION
# =====================================================

def polynomial_mutation(pop):

    mutation_mask = np.random.rand(*pop.shape) < mutation_rate
    r = np.random.rand(*pop.shape)

    delta = np.where(
        r < 0.5,
        (2*r)**(1/(eta_m+1)) - 1,
        1 - (2*(1-r))**(1/(eta_m+1))
    )

    pop[mutation_mask] += delta[mutation_mask] * (Xu - Xl)

    return np.clip(pop, Xl, Xu)

# =====================================================
# INITIAL POPULATION
# =====================================================

population = np.random.rand(pop_size, n_jobs)

# =====================================================
# GA LOOP
# =====================================================

best_solution = None
best_fitness = -1e18

for gen in range(generations):

    fit = fitness(population)

    best_idx = np.argmax(fit)

    if fit[best_idx] > best_fitness:
        best_fitness = fit[best_idx]
        best_solution = population[best_idx].copy()

    if gen % 50 == 0:
        print(f"Generation {gen}, Best Fitness: {best_fitness}")

    selected = selection(population, fit)
    offspring = sbx_crossover(selected)
    population = polynomial_mutation(offspring)


final_assignment = decode(best_solution.reshape(1, -1))[0]

used_resource = np.zeros(n_machines)
np.add.at(
    used_resource,
    final_assignment,
    resource[final_assignment, np.arange(n_jobs)]
)

print("\nFinal Total Cost:", 
      cost[final_assignment, np.arange(n_jobs)].sum())

print("Capacity Used:", used_resource)
print("Capacity Limit:", capacity)
print("Capacity Violation:", np.maximum(used_resource - capacity, 0))
print("Is Feasible?:", np.all(used_resource <= capacity))
print("\nFinal Assignment (job → machine):")
print(final_assignment)
```
